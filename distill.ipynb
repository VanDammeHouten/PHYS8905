```python
import matplotlib.pyplot as plt
import numpy as np

# Create a vector of theta values, with +1e-16 to prevent log of zero later.
theta = np.arange(0., np.pi/4, np.pi/400)+1e-16
cs=np.cos(theta)**2
sn=np.sin(theta)**2
# The entanglement of the initial state.
ent0=-3*(cs*np.log(cs)+sn*np.log(sn))/np.log(2)
# The average distilled entanglement
ent1=2*sn**3 + 3*cs*sn*(2*sn+np.log(3)/np.log(2))
plt.plot(theta,ent0,theta,ent1)
plt.xlabel(r'$\theta$')
plt.ylabel('entanglement')
plt.legend(['initial entanglement', 'final entanglement'])
```

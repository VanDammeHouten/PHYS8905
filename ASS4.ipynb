The CPTP map between a system $\mathcal{H}_Q$ and an environment $\mathcal{H}_E$ for a density operator $\rho_E$ acted on $\mathcal{H}_Q\otimes\mathcal{H}_E$ by a unitary operator $U$ can be expressed by:
\begin{aligned}
    M(\rho) &= \mathrm{tr}_{\mathcal{H}_E}(U(\rho\otimes\rho_E)U^{\dagger})
\end{aligned}

$\mathcal{H}_E$ has the basis $\ket{0}_E,\ket{1}_E,\ket{2}_E,\ket{3}_E$ and our system interacts with the environment according to the unitary $U_{QE}$. The partial trace over the environment can be reexpressed in its basis giving the CPTP map between a system and environment acted on by a unitary $U_{QE} as:
\begin{aligned}
    M(\rho) &= \sum_{k=0}^4 {}_{E}\bra{k}U_{QE}(\rho\otimes\rho_E)U_{QE}^{\dagger}\ket{k}_E
\end{aligned}

```python
import qutip as qt
import numpy as np
z0 = qt.basis(2,0)
z1 = qt.basis(2,1)
# Define observables for Alice and Bob.
B1 = (qt.sigmaz() + qt.sigmax()) / np.sqrt(2)
B2 = (qt.sigmaz() - qt.sigmax()) / np.sqrt(2)
A1 = qt.sigmaz()
A2 = qt.sigmax()
# Question 3, part (a)
# Find eigenvalues of Bob's observables.
print('part (a)')
print('eigenvalues of B1: ', B1.eigenenergies())
print('eigenvalues of B2: ', B2.eigenenergies())
print()
# Question 3, part (b)
# Use the standard formula for the CHSH operator for S:
S = qt.tensor(A1,B1) + qt.tensor(A2,B1) + qt.tensor(A1,B2) - qt.tensor(A2,B2)
print('part (b)')
print('CHSH operator: ', S)
print('Eigenvalues of CHSH operator: ', S.eigenenergies())
# The maximum value of the CHSH quantity is the maximum expectation
# value of this operator, which is the maximum eigenvalue.
print('Maximum of CHSH quantity: ', max(S.eigenenergies()))
# This is actually 2*sqrt(2).
print('2*sqrt(2): ', 2*np.sqrt(2))
```

```python
# Question 3, part (c)
print('part (c)')
# This gives all the eigenvalues and eigenvectors.
print('All eigenvectors and eigenvalues: ', S.eigenstates())
print()
# This gives all eigenvectors.
print('All eigenvectors: ', S.eigenstates()[1])
print()
# This gives the eigenvector corresponding to the largest eigenvalue, which is the last listed.
print('The eigenvector we want: ', S.eigenstates()[1][-1])
print()
# You will see that this is the equal superposition of 00 and 11, except for an unobservable global phase:
zer = qt.qstate('d')
one = qt.qstate('u')
print('The Bell state: ', (qt.tensor(zer,zer)+qt.tensor(one,one))/np.sqrt(2))
# This means that these measurements will maximise violation of the BEll inequality for the standard Bell state.
```
